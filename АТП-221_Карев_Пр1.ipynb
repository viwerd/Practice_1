```python
import pandas as pd # Подключаем библиотеку pandas для анализа данных
animals_data = {'Name':['Cat','Wolf','Tiger','Elephant'],
              'Species':['Carnivore','Carnivore','Carnivore','Herbivore'],
              'Weight':[4 , 60 , 180 , 600],
              'Count':[5 , 10 , 4 , 2]} # Словарь
columns = ['Name','Species','Weight','Count'] # Название колонок
df = pd.DataFrame(animals_data , columns = columns) # Создание датафрейма
df.to_csv('animals.csv', index=False) # Экспорт данных
N = 2
df
```

```python
# Убираем из словаря ноль (https://pythonpip.ru/pandas/pandas-fillna)
df.fillna(value = 0)
df.dropna(1)
```
